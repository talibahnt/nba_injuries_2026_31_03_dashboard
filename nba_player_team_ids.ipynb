```python
#Use the Hidden JSON API directly
#This is cleaner and faster than scraping HTML. The G League stats site calls stats.gleague.nba.com/stats/ endpoints 
import requests
import pandas as pd

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 Chrome/120.0.0.0 Safari/537.36",
    "Referer": "https://stats.gleague.nba.com/",
    "Accept": "application/json, text/plain, */*",
    "Accept-Language": "en-US,en;q=0.9",
    "Origin": "https://stats.gleague.nba.com",
    "x-nba-stats-origin": "stats",
    "x-nba-stats-token": "true",
}

# --- Player Stats ---
def get_player_stats(season="2024-25", season_type="Regular Season"):
    url = "https://stats.gleague.nba.com/stats/leaguedashplayerstats"
    params = {
        "Season": season,
        "SeasonType": season_type,
        "PerMode": "PerGame",
        "MeasureType": "Base",
        "LeagueID": "20",   # 20 = G League, 00 = NBA
        "DateFrom": "",
        "DateTo": "",
        "GameScope": "",
        "GameSegment": "",
        "LastNGames": 0,
        "Location": "",
        "Month": 0,
        "OpponentTeamID": 0,
        "Outcome": "",
        "PORound": 0,
        "PaceAdjust": "N",
        "Period": 0,
        "PlayerExperience": "",
        "PlayerPosition": "",
        "PlusMinus": "N",
        "Rank": "N",
        "StarterBench": "",
        "TeamID": 0,
        "TwoWay": 0,
        "VsConference": "",
        "VsDivision": "",
    }
    response = requests.get(url, headers=headers, params=params)
    print(f"Status: {response.status_code}")

    data = response.json()
    cols = data["resultSets"][0]["headers"]
    rows = data["resultSets"][0]["rowSet"]
    return pd.DataFrame(rows, columns=cols)

# --- Team Stats ---
def get_team_stats(season="2024-25", season_type="Regular Season"):
    url = "https://stats.gleague.nba.com/stats/leaguedashteamstats"
    params = {
        "Season": season,
        "SeasonType": season_type,
        "PerMode": "PerGame",
        "MeasureType": "Base",
        "LeagueID": "20",
        "DateFrom": "",
        "DateTo": "",
        "GameScope": "",
        "GameSegment": "",
        "LastNGames": 0,
        "Location": "",
        "Month": 0,
        "OpponentTeamID": 0,
        "Outcome": "",
        "PORound": 0,
        "PaceAdjust": "N",
        "Period": 0,
        "PlusMinus": "N",
        "Rank": "N",
        "TwoWay": 0,
        "VsConference": "",
        "VsDivision": "",
    }
    response = requests.get(url, headers=headers, params=params)
    print(f"Status: {response.status_code}")

    data = response.json()
    cols = data["resultSets"][0]["headers"]
    rows = data["resultSets"][0]["rowSet"]
    return pd.DataFrame(rows, columns=cols)

# --- Standings ---
def get_standings(season="2024-25"):
    url = "https://stats.gleague.nba.com/stats/leaguestandingsv3"
    params = {
        "LeagueID": "20",
        "Season": season,
        "SeasonType": "Regular Season",
        "SeasonYear": "",
    }
    response = requests.get(url, headers=headers, params=params)
    print(f"Status: {response.status_code}")

    data = response.json()
    cols = data["resultSets"][0]["headers"]
    rows = data["resultSets"][0]["rowSet"]
    return pd.DataFrame(rows, columns=cols)


# Run them
players_df = get_player_stats()
print(players_df[["PLAYER_NAME", "TEAM_ABBREVIATION", "GP", "PTS", "REB", "AST"]].head(10))

teams_df = get_team_stats()
print(teams_df[["TEAM_NAME", "GP", "W", "L", "PTS", "REB", "AST"]].head(10))

standings_df = get_standings()
print(standings_df.head(10))

# Save to CSV
players_df.to_csv("gleague_player_stats.csv", index=False)
teams_df.to_csv("gleague_team_stats.csv", index=False)
standings_df.to_csv("gleague_standings.csv", index=False)
print("All files saved!")
```
